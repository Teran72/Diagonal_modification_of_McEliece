```python
def get_cycles(n, p, m):
    l = [(p * i) % n for i in range(n)]
    return l
```

```python
def color_table(n, p, m):
    l = get_cycles(n, p, m)
    colors = [[-1 for i in range(n)] for j in range(n)]
    cnt = 1
    for i in range(n):
        for j in range(n):
            if (i == j):
                continue
            if (colors[i][j] >= 0):
                continue
            pi = i
            pj = j
            while (colors[pi][pj] < 0):
                colors[pi][pj] = cnt
                colors[pj][pi] = cnt
                pi = l[pi]
                pj = l[pj]
            cnt += 1
    return colors
```
